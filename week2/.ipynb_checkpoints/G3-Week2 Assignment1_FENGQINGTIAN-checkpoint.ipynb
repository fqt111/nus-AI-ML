## Q3_A

```python
import numpy as np
import matplotlib.pyplot as plt

#raw data
x=np.array([0,4,8,12,16,20,24,28,32])
y=np.array([394.33, 329.50, 291.00, 255.17, 229.33, 204.83, 179.00, 163.83, 150.33])
```

we use the function to calculate$\beta$,$\beta=（X^{'}X）^{-1}X^{'}Y$
1. Z=$（X^{'}X）$
2. Zinv=$Z^{-1}$
3. Z2=$ZinvX^{'}$
4. Z3=$Z2Y$

```python
Xt=np.array([np.ones(9),x])
X=Xt.transpose()
Z = np.matmul(Xt,X)
Zinv=np.linalg.inv(Z) 
Z2=np.matmul(Zinv,Xt) 
Z3=np.matmul(Z2,y)
```

```python
# the expression for the best fit line
e1= Z3[0]
beta= Z3[1]
ypred=beta*x+e1
print("Q3_A the best fit line: ""y=",beta,"* x+",e1)

# error performance on the training dataset
def calculateMSE(X,Y,m,b):
    return sum([(y-m*x -b)**2 for x,y in zip(X,Y)])/len(X)
print("the error performance",calculateMSE(x,y,beta,e1))
```

## Q3_B

```python
# image
fig, ax = plt.subplots(1, 1, figsize=(9, 5))
plt.plot(x,y,'Dy')
plt.plot(x,ypred,'g')
plt.xlabel("x:mileage")
plt.ylabel("y:Tire Tread Depth(in mile)")
plt.title("Q3_B")
plt.grid()
```

## Q3_C

```python
# take the first 8 elements
x2 = x[0:8]
y2 = y[0:8]

# find the fit line
Xt = np.array([np.ones(8),x2])
X=Xt.transpose()
Z = np.matmul(Xt,X)
Zinv=np.linalg.inv(Z)
Z2=np.matmul(Zinv,Xt)
Z3=np.matmul(Z2,y2)

# result

e2 = Z3[0]
beta2 = Z3[1]
ypred2=beta2*x2+e2
print("Q3_C: ""y=",beta2,"* x+",e2)
```

```python
# error performance
print("the test error performance: ",calculateMSE(x2,y2,beta2,e2))
print("the train error performance: ",(beta2*x[8]+e2-y[8])**2)

# image
fig, ax = plt.subplots(1, 1, figsize=(9, 5))
plt.plot(x,y,'oc')
plt.plot(x[0:8],ypred2,'k')
plt.xlabel("x:mileage")
plt.ylabel("y:Tire Tread Depth(in mile)")
plt.title("Q3_C")
plt.grid()
```

## Q3_D

```python
import sklearn.linear_model as lm
lr = lm.LinearRegression()
lr.fit(x[:, np.newaxis],y)
x_tr = np.linspace(0., 32, 50)
ypred3 = lr.predict(x_tr[:, np.newaxis])
beta3 = lr.coef_[0]
e3 = lr.intercept_
print("Q3_D the best fit line(LinearRegression): ""y=",beta3,"* x+",e3)
```

```python
ridge = lm.RidgeCV()
ridge.fit(x[:, np.newaxis],y)
x_tr2 = np.linspace(0., 32, 50)
ypred4 = ridge.predict(x_tr2[:, np.newaxis])
beta4 = ridge.coef_[0]
e4 = ridge.intercept_
print("Q3_D the best fit line(RidgeCV): ""y=",beta4,"* x+",e4)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))
ax.plot(x_tr, ypred3,'m',label='LinearRegression')
ax.plot(x_tr2, ypred4,'y--',label='RidgeCV')
ax.plot(x, y, 'or')
ax.set_title("Q3_D")
plt.legend(loc=1)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

As we can see the Least_square_9 and skearn_LinearRegression whose coefficients are the same. Then we can guess the API of skearn.LinearRegression has also uses method of Linear regression based on least squares.
Besides,we also use Ridge regression which is a dedicated to a total of linear biased estimation of regression data analysis method. In essense, it is an improved least squares estimation method. By giving up the unbiasedness of traditional least-square method, this method is obtained at the cost of partial information loss and precision reduction with more realistic and reliable regression coefficients, and the fitting of ill-conditioned data is stronger than the least square method. So the coefficients of them have a few differences.
